```python
# define constants
input_dataset_parent_path = '/home/vamshikrishnadude/Downloads/Driving-range-Prediction-and-Energy-Consumption-in-EV-s/input_dataset/'
path_to_preprocessed_with_outliers_csv = input_dataset_parent_path+ 'final_preprocessed_with_outliers.csv'
path_to_preprocessed_without_outliers_csv = input_dataset_parent_path+ 'final_preprocessed_without_outliers.csv'
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv(path_to_preprocessed_with_outliers_csv)

Y = df['trip_distance(km)']
X = df.drop(columns=['trip_distance(km)'])

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print("Shape of X train : ", X_train.shape)
print("Shape of X test : ", X_test.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
# city
plt.figure(figsize=(20, 10))
plt.suptitle("Countplots")
plt.subplot(2, 3, 1)
plt.title("City counts")
sb.countplot(X_train['city'])

plt.subplot(2, 3, 2)
plt.title("Motor way counts")
sb.countplot(X_train['motor_way'])

plt.subplot(2, 3, 3)
plt.title("country_roads counts")
sb.countplot(X_train['country_roads'])

plt.subplot(2, 3, 4)
plt.title("driving_style counts")
sb.countplot(X_train['encoded_driving_style'])

plt.subplot(2, 3, 5)
plt.title("A/C counts")
sb.countplot(X_train['A/C'])

plt.subplot(2, 3, 6)
plt.title("park_heating counts")
sb.countplot(X_train['park_heating'])

plt.show()
```
