```python
# define constants

input_dataset_parent_path = '/home/vamshikrishnadude/Downloads/Driving data prediction/input_dataset/'
path_to_volks = input_dataset_parent_path + 'volkswagen_e_golf.csv'
path_to_mits = input_dataset_parent_path +'/mitsubishi_imiev.csv'

path_to_merged_csv = input_dataset_parent_path+ 'data.csv'
```

```python
# import libraries
import pandas as pd
import os
from IPython.core.display import display
```

```python
# Global variables
df = pd.DataFrame();
```

```python
# Check the data files exists and load in to df

if os.path.exists(path_to_merged_csv) :
    df = pd.read_csv(path_to_merged_csv);
    print("Merged data file exists. loaded.")
    display(df.head(5))
else :
    # read both test files and append.
    print("Merged data file doesnot exist")
    print('='*100)
    if os.path.exists(path_to_volks) :
        df = pd.read_csv(path_to_volks)
        print("Volkswagen data file exists. loaded.")
        display(df.head(5))
        print("Shape of Volkswagen data : ", df.shape)
    if os.path.exists(path_to_mits) :
        print('='*100)
        mits_df = pd.read_csv(path_to_mits, encoding='latin-1')
        print("mitsubishi_imiev data file exists. loaded.")
        display(mits_df.head(5))
        print("Shape of mitsubishi_imiev data : ", mits_df.shape)
        df = df.append(mits_df)
        print('='*100)
        print('Merged Data file : ')
        display(df.head(5))
        df.to_csv(path_to_merged_csv)
        print("Shape of Merged data : ", df.shape)
```
